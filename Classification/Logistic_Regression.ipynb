```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#exp() to calculate the value of e^z
input_val = 1
exp_val = np.exp(input_val)
print("Input to exp: ", input_val)
print("Output of exp: ", exp_val)

input_arr = np.array([1, 2, 3])
exp_arr = np.exp(input_arr)
print("Input to exp: ", input_arr)
print("Output of exp: ", exp_arr)
```

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
```

```python
z_tmp = np.arange(-10, 11)
y = sigmoid(z_tmp)
#pretty printing of arrays next to each other
np.set_printoptions(precision=3)
print("Input (z), Output (sigmoid(z))")
print(np.c_[z_tmp, y])
```

```python
plt.figure()
plt.plot(z_tmp, y, marker='o')
plt.xlabel("z")
plt.ylabel("sigmoid(z)")
plt.title("Logistic Regression (Sigmoid Function)")
plt.grid(True)
plt.show()
```
